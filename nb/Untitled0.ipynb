```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
from scipy.interpolate import Rbf, InterpolatedUnivariateSpline
```

```python
data = np.loadtxt('../data/data.csv', delimiter=',')
```

```python
data.shape
```

```python
t = np.linspace(0, 336, 337).astype(int)
rcParams['figure.figsize'] = 18, 10
p1, = plot(t, data[:, 0], 'r')
p2, = plot(t, data[:, 1], 'b')
p3, = plot(t, data[:, 2], 'g')
p4, = plot(t, data[:, 3], 'm')
p5, = plot(t, data[:, 4], 'y')
legend([p5, p4, p3, p2, p1], ["Contract 4", "Contract 3", "Contract 2", "Contract 1", "Oil Price"])
plt.xlabel('No. of Months from Jan 1986')
plt.ylabel('Price in USD per Barrel')
plt.title('Crude Oil Price and Futures')
plt.show()
```

```python
db7 = pywt.Wavelet('db7')
```

```python
cA5, cD5, cD4, cD3, cD2, cD1 = pywt.wavedec(data[:, 0], db7, level=5)
```

```python
# Extend the samples of cA5 using an interpolated univariate spline. This, by fitting this spline, we extended the 
# data to values [0:349] rather than [0:336]
t0 = np.linspace(0,336,cA5.shape[0]).astype(int)
ius_cA5 = InterpolatedUnivariateSpline(t0[:-2],cA5[:-2])
ti = np.linspace(0,336,cA5.shape[0]).astype(int)
cA5i1 = ius_cA5(ti)
Wi1 = pywt.waverec([cA5i1, cD5, cD4, cD3, cD2, cD1], 'db7')

rbf_cA5 = Rbf(t0,cA5)
cA5i2 = rbf_cA5(ti)
Wi2 = pywt.waverec([cA5i2, cD5, cD4, cD3, cD2, cD1], 'db7')

rcParams['figure.figsize'] = 18, 10
p1, = plot(t, data[:, 0], 'r')
#p2, = plot(Wi1, 'b')
p3, = plot(Wi2, 'g')

legend([p3, p2, p1], ["RBF method", "Univariate Spline", "Oil Price"])
#plt.xlabel('No. of Months from Jan 1986')
#plt.ylabel('Price in USD per Barrel')
#plt.title('Crude Oil Price and Futures')
plt.show()
```
