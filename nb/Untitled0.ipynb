```python
db7 = pywt.Wavelet('db7')
```

```python
cA5, cD5, cD4, cD3, cD2, cD1 = pywt.wavedec(data[:, 0], db7, level=5)
```

## Forecasting

```python
t0 = np.linspace(0,336,cA5.shape[0]).astype(int)
ius_cA5 = InterpolatedUnivariateSpline(t0[:-2],cA5[:-2])
ti = np.linspace(0,336,cA5.shape[0]).astype(int)
cA5i1 = ius_cA5(ti)

datai1 = pywt.waverec([cA5i1, cD5, cD4, cD3, cD2, cD1], 'db7')
```

```python
rbf_cA5 = Rbf(t0[:-2],cA5[:-2])
cA5i2 = rbf_cA5(ti)
datai2 = pywt.waverec([cA5i2, cD5, cD4, cD3, cD2, cD1], 'db7')
```

```python
rcParams['figure.figsize'] = 18, 10
p1, = plot(t, data[:, 0], 'r')
p2, = plot(datai1, 'b')
p3, = plot(datai2, 'g')

legend([p3, p2, p1], ["RBF method", "Univariate Spline", "Oil Price"])

plt.show()
```
